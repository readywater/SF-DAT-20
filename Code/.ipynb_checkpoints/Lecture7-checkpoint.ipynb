### Single Imputation - Regression Lines

```python
# Now let's run a model and see what is the best way to predict age
TitanicRawData.corr()
```

```python
from sklearn.linear_model import LinearRegression
url = "https://raw.githubusercontent.com/ga-students/SF-DAT-20/master/Data/titanic.csv"
TitanicRawData = pd.read_csv(url)
TitanicDroppedValues = TitanicRawData.dropna(subset = ['age'])
```

```python
linreg = LinearRegression()
X = TitanicDroppedValues[['pclass','sibsp','fare','survived']]
y = TitanicDroppedValues['age']
linreg.fit(X,y)
```

```python
from sklearn import feature_selection
pvals = feature_selection.f_regression(X,y)[1] #That's how we extract p-values
print(pvals)
```

```python
X = TitanicRawData[['pclass','sibsp','fare','survived']]
TitanicRawData['age_predict'] = linreg.predict(X)
TitanicRawData['age'].fillna(value = TitanicRawData['age_predict'], inplace = True)
```

#### Single Imputation Regression line with Error

```python
url = "https://raw.githubusercontent.com/ga-students/SF-DAT-20/master/Data/titanic.csv"
TitanicRawData = pd.read_csv(url)
TitanicDroppedValues = TitanicRawData.dropna(subset = ['age'])

linreg = LinearRegression()

X = TitanicDroppedValues[['pclass','sibsp','fare','survived']]
y = TitanicDroppedValues['age']
linreg.fit(X,y)
y_hat = linreg.predict(X)
Standard_error = (sum((y-y_hat)**2)/(len(y)-2))**.5
print(Standard_error)

```

```python
y_min = y.min()
y_max = y.max()
X = TitanicRawData[['pclass','sibsp','fare','survived']]
print(len(TitanicRawData))
#let's add error term to our prediction
TitanicRawData['predicted_age_error'] = linreg.predict(X) + np.random.normal(0,Standard_error,len(TitanicRawData))


print(TitanicRawData['predicted_age_error'].min())

# we don't want to predict below min
for i in range(0,(len(TitanicRawData))):
    if (TitanicRawData.loc[i,'predicted_age_error'] < y.min()):
        TitanicRawData.loc[i,'predicted_age_error'] = y_min
print(TitanicRawData['predicted_age_error'].min())


# we don't want to predict above max
for i in range(0,(len(TitanicRawData))):
    if (TitanicRawData.loc[i,'predicted_age_error'] > y.max()):
        TitanicRawData.loc[i,'predicted_age_error'] = y_max


TitanicRawData['age'].fillna(value = TitanicRawData['predicted_age_error'], inplace = True)


TitanicRawData.head(20)
```
